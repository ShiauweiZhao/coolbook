```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

import pandas as pd
import matplotlib.pyplot as plt
import pytz

# 读取CSV文件
df = pd.read_csv('2034.csv')

# 将 createTime 列转换为 datetime 类型（以毫秒为单位）
df['createTime'] = pd.to_datetime(df['createTime'], unit='ms')

# 设置时区为 UTC
df['createTime'] = df['createTime'].dt.tz_localize('UTC')

# 转换为 UTC+8 时区
df['createTime'] = df['createTime'].dt.tz_convert(pytz.timezone('Asia/Shanghai'))

# 提取时分秒
df['timeHour'] = df['createTime'].dt.strftime('%H:%M:%S')
print(df['timeHour'])

# 按照sn分组
grouped = df.groupby('sn')

# 创建图形和子图对象
fig, ax = plt.subplots()

# 遍历每个分组，绘制时间序列图
for name, group in grouped:
    ax.plot(group['timeHour'], group['starNum'], label=name)

# 设置标题和标签
ax.set_title('Star Number Over Time')
ax.set_xlabel('Time')
ax.set_ylabel('Star Number')

# 移除图例
ax.legend().remove()

# 显示图形
plt.show()

```

```python
import numpy as np
# 按照sn分组
grouped = df.groupby('sn')

# 创建图形和子图对象
fig, ax = plt.subplots()

# 遍历每个分组，绘制时间序列图
for name, group in grouped:
    ax.plot(np.array(group['timeHour'])[:, np.newaxis], np.array(group['fixMode'])[:, np.newaxis], label=name)

# 设置标题和标签
ax.set_title('Fix Mode Over Time')
ax.set_xlabel('Time')
ax.set_ylabel('Fix Mode')

# 移除图例
ax.legend().remove()

# 显示图形
plt.show()
```

```python
# 设置时间范围
start_time = pd.to_datetime('19:00:00', format='%H:%M:%S')
end_time = pd.to_datetime('21:30:00', format='%H:%M:%S')

# 筛选出在时间范围内的数据
filtered_data = df[(df['timeHour'] >= start_time) & (df['timeHour'] <= end_time)]

# 找到定位模式发生跳变的SN
sn_with_mode_change = filtered_data['sn'][filtered_data['fixMode'].diff() != 0]


# 找到 'fixMode' 或 'starNum' 列中值为0的行
zero_fixMode_or_starNum_rows = filtered_data[(filtered_data['fixMode'] == 0) | (filtered_data['starNum'] == 0)]

# 输出这些行的 'sn' 值并去重
unique_zero_sn = zero_fixMode_or_starNum_rows['sn'].unique().tolist()

# 输出unique_zero_sn
print(unique_zero_sn)
```

```python
# 计算每种定位模式的出现次数
fixMode_counts = df['fixMode'].value_counts()

# 绘制条形图
fixMode_counts.plot(kind='bar', title='Fix Mode Counts')

# 显示图形
plt.show()
```

```python

sn_to_plot ='B01Z01TJDEAA13'
# 筛选出指定SN的数据
sn_data = df[df['sn'] == sn_to_plot]

# 获取星数和定位模式数据
starNum = sn_data['starNum']
fixMode = sn_data['fixMode']

# 创建图形和子图对象
fig, ax1 = plt.subplots()

# 绘制星数图
ax1.plot(starNum, color='blue', label='StarNum')
ax1.set_ylabel('StarNum')
ax1.set_xlabel('Time')

# 创建第二个y轴对象
ax2 = ax1.twinx()

# 绘制定位模式图
ax2.plot(fixMode, color='red', label='FixMode')
ax2.set_ylabel('FixMode')

# 添加图例
ax1.legend(loc='upper left', bbox_to_anchor=(0, 1))
ax2.legend(loc='upper right', bbox_to_anchor=(1, 1))

# 设置标题
plt.title('StarNum and FixMode for SN {sn_to_plot}')

# 显示图形
plt.show()
```
